```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
```

```python
karate_graph = nx.read_gml(r"C:\Users\pc\Downloads\ADS Lab programs\karate.gml")
print("Basic Information about the Karate Club Network:")
print(karate_graph)
```

```python
actor_metadata = {}
for node in karate_graph.nodes:
    actor_metadata[node] = karate_graph.nodes[node]
```

```python
degree_centrality = nx.degree_centrality(karate_graph)
betweenness_centrality = nx.betweenness_centrality(karate_graph)
closeness_centrality = nx.closeness_centrality(karate_graph)
eigenvector_centrality = nx.eigenvector_centrality(karate_graph)
pagerank_centrality = nx.pagerank(karate_graph)
```

```python
print("\nCentrality Measures:")
for node in karate_graph.nodes:
    print(f"Node {node}:")
    print(f"  Degree Centrality: {degree_centrality[node]}")
    print(f"  Betweenness Centrality: {betweenness_centrality[node]}")
    print(f"  Closeness Centrality: {closeness_centrality[node]}")
    print(f"  Eigenvector Centrality: {eigenvector_centrality[node]}")
    print(f"  PageRank Centrality: {pagerank_centrality[node]}")
    print()
```

```python
k_components = nx.k_components(karate_graph)
clustering_coefficient = nx.average_clustering(karate_graph)

print("\nK-components:")
print(k_components)
print("\nClustering Coefficient:")
print(clustering_coefficient)
```

```python
communities_girvan_newman = tuple(sorted(c) for c in community.girvan_newman(karate_graph))

print("\nCommunities using Girvan-Newman algorithm:")
print(communities_girvan_newman)
```

```python
communities_louvain = nx.community.greedy_modularity_communities(karate_graph)

print("\nCommunities using Louvain method:")
print(communities_louvain)
```

```python
nx.draw(karate_graph, with_labels=True)
plt.show()
```
